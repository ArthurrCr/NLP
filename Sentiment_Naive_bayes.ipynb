# Probability and Bayes’ Rule

Para começar, vamos revisar o que são probabilidades e probabilidades condicionais, como elas operam e como podem ser expressas matematicamente. Em seguida, abordarei como derivar a regra de Bayes a partir da definição de probabilidades condicionais. A regra de Bayes é aplicada em muitos campos diferentes, desde a medicina até a educação, e é amplamente utilizada no PLN. Uma vez que você entenda a teoria por trás da regra de Bayes, poderá usá-la para realizar a análise de sentimento em tweets, que será a tarefa desta semana.

Imagine que você tem um corpus extenso de tweets que podem ser categorizados como tendo sentimento positivo ou negativo, mas não ambos. Dentro desse corpus, a palavra "feliz" às vezes é rotulada como positiva e às vezes como negativa. Vamos explorar por que essa situação ocorre. Uma maneira de pensar sobre probabilidades é contar com que frequência os eventos ocorrem. Suponha que você defina o evento A como um tweet sendo rotulado como positivo, então a probabilidade do evento A, mostrada como P(A) aqui, é calculada como a razão entre a contagem de tweets positivos no corpus dividida pelo número total de tweets no corpus. Neste exemplo, esse número é 13 em 20, ou 0,65. Você também pode expressar esse valor como uma porcentagem, 65% de positivos.



Vale a pena notar que a probabilidade complementar aqui, que é a probabilidade dos tweets expressarem um sentimento negativo, é apenas igual a um menos a probabilidade de um sentimento positivo. Observe que, para isso ser verdadeiro, todos os tweets devem ser categorizados como positivos ou negativos, mas não ambos.



Vamos definir o Evento B de maneira semelhante, contando os tweets que contêm a palavra "feliz". Nesse caso, o número total de tweets contendo a palavra "feliz", mostrado aqui como N_feliz, é 4. Aqui está outra maneira de ver isso. Observe a seção do diagrama onde os tweets são rotulados como positivos e também contêm a palavra "feliz". 



No contexto deste diagrama, a probabilidade de um tweet ser rotulado como positivo e também conter a palavra "feliz" é apenas a razão entre a área da interseção dividida pela área do corpus inteiro. Em outras palavras, se houver 20 tweets no corpus e três deles forem rotulados como positivos e também contiverem a palavra "feliz", então a probabilidade associada é 3 dividido por 20 ou 0,15.

# Bayes’ Rule

Vamos examinar as probabilidades condicionais para nos ajudar a entender a regra de Bayes. Em outras palavras, se eu lhe disser para adivinhar como está o tempo dado que estamos na Califórnia e é inverno, você terá uma ideia muito melhor do que se eu apenas pedisse para adivinhar como está o tempo. Para derivar a regra de Bayes, vamos primeiro observar as probabilidades condicionais.

Agora pense no que acontece se, em vez de considerar todo o corpus, você considerar apenas os tweets que contêm a palavra "feliz". Isso é o mesmo que dizer: dado que um tweet contém a palavra "feliz", você estaria considerando apenas os tweets dentro do círculo azul, onde muitos dos tweets positivos agora estão excluídos. Nesse caso, a probabilidade de um tweet ser positivo, dado que contém a palavra "feliz", simplesmente se torna o número de tweets que são positivos e também contêm a palavra "feliz". Dividimos isso pelo número de tweets que contêm a palavra "feliz". Como você pode ver por esse cálculo, seu tweet tem uma probabilidade de 75% de ser positivo se contiver a palavra "feliz".



Você poderia fazer o mesmo caso para tweets positivos. A área roxa denota a probabilidade de que um tweet positivo contenha a palavra "feliz". Nesse caso, a probabilidade é 3 em 13, que é 0,231. Com toda essa discussão sobre a probabilidade de faltar certas condições, estamos falando sobre probabilidades condicionais. Probabilidades condicionais podem ser interpretadas como a probabilidade de um resultado B sabendo que o evento A já aconteceu, ou dado que estou olhando para um elemento do conjunto A, a probabilidade de que ele também pertença ao conjunto B. Aqui está outra maneira de olhar para isso com um diagrama de Venn que você viu antes.



Usando o exemplo anterior, a probabilidade de um tweet ser positivo, dado que tem a palavra "feliz", é igual à probabilidade da interseção entre os tweets que são positivos e os tweets que têm a palavra "feliz" dividida pela probabilidade de um tweet do corpus ter a palavra "feliz". Vamos dar uma olhada mais de perto na equação do slide anterior. Você pode escrever uma equação semelhante simplesmente trocando a posição das duas condições. Agora, você tem a probabilidade condicional de um tweet conter a palavra "feliz", dado que é um tweet positivo. Armado com ambas as equações, você está pronto para derivar a regra de Bayes.



Para combinar essas equações, observe que a interseção representa a mesma quantidade, não importa como está escrita. Sabendo disso, você pode removê-la da equação. Com um pouco de manipulação algébrica, você pode chegar a esta equação. 



Esta é agora uma expressão da regra de Bayes no contexto do problema de análise de sentimento anterior. Mais geralmente, a regra de Bayes afirma que a probabilidade de x dado y é igual à probabilidade de y dado x vezes a razão entre a probabilidade de x sobre a probabilidade de y.



Suponha que em seu conjunto de dados, 25% dos tweets positivos contenham a palavra ‘feliz’. Você também sabe que um total de 13% dos tweets no seu conjunto de dados contém a palavra “feliz” e que 40% do número total de tweets são positivos. Você observa o tweet: ''feliz em aprender PNL'. Qual é a probabilidade de este tweet ser positivo?


$$ P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)} $$

Onde:
- $ P(A|B) $ é a probabilidade de que um tweet seja positivo dado que contém a palavra "happy".
- $ P(B|A) $ é a probabilidade de que um tweet contenha a palavra "happy" dado que é positivo.
- $ P(A) $ é a probabilidade de que um tweet seja positivo.
- $ P(B) $ é a probabilidade de que um tweet contenha a palavra "happy".

De acordo com o problema, temos os seguintes dados:
- $ P(B|A) = 0.25 $ (25% dos tweets positivos contêm a palavra "happy")
- $ P(B) = 0.13 $ (13% dos tweets contêm a palavra "happy")
- $ P(A) = 0.40 $ (40% dos tweets são positivos)

Substituindo esses valores na fórmula da regra de Bayes, obtemos:

$$ P(A|B) = \frac{0.25 \cdot 0.40}{0.13} $$

Vamos calcular isso passo a passo:

$$ P(A|B) = \frac{0.25 \cdot 0.40}{0.13} $$
$$ P(A|B) = \frac{0.10}{0.13} $$
$$ P(A|B) \approx 0.7692 $$

Portanto, a probabilidade de que o tweet "happy to learn NLP" seja positivo, dado que contém a palavra "happy", é aproximadamente 76.92%.

# Naïve Bayes Introduction

Naive Bayes é um exemplo de aprendizado de máquina supervisionado e compartilha muitas semelhanças com o método de regressão logística que você utilizou nas tarefas anteriores. É chamado de "naive" (ingênuo) porque esse método faz a suposição de que as características que você está usando para a classificação são todas independentes. Na realidade, isso raramente é o caso. No entanto, como você verá, ele ainda funciona bem como um método simples para análise de sentimentos.

**Preparação dos Dados:**
Assim como antes, você começará com dois corpora: um para os tweets positivos e outro para os tweets negativos. Você precisa extrair o vocabulário ou todas as diferentes palavras que aparecem em seu corpus, junto com suas contagens. Você obtém as contagens de palavras para cada ocorrência de uma palavra no corpus positivo e faz o mesmo para o corpus negativo.

**Contagem Total de Palavras:**
Depois, você obterá uma contagem total de todas as palavras em seu corpus positivo e fará o mesmo para seu corpus negativo. Isso significa que você está apenas somando as linhas desta tabela. Por exemplo, para tweets positivos, há um total de 13 palavras e para tweets negativos, um total de 12 palavras.



**Cálculo das Probabilidades Condicionais:**
Este é o primeiro novo passo para Naive Bayes e é muito importante porque permite calcular as probabilidades condicionais de cada palavra, dado a classe a que pertence. Agora, divida a frequência de cada palavra em uma classe pela soma correspondente de palavras na classe. Por exemplo, para a palavra "eu", a probabilidade condicional para a classe positiva seria $ \frac{3}{13} $. Você armazena esse valor em uma nova tabela com o valor correspondente, 0.24. Para a palavra "eu" na classe negativa, você obtém $ \frac{3}{12} $ e armazena isso em sua nova tabela com o valor correspondente, 0.25.

**Construção da Tabela de Probabilidades:**
Agora, aplique o mesmo procedimento para cada palavra em seu vocabulário para completar a tabela de probabilidades condicionais. Uma propriedade chave desta tabela é que se você somar todas as probabilidades para cada classe, você obterá 1.



**Análise da Tabela de Probabilidades:**
Vamos investigar essa tabela mais a fundo para entender o significado desses números. Primeiramente, observe quantas palavras têm uma probabilidade condicional quase idêntica, como "estou aprendendo" e "NLP". O interessante aqui é que palavras que são igualmente prováveis não adicionam nada ao sentimento. Em contraste com essas palavras neutras, observe algumas outras palavras como "feliz", "triste" e "não". Elas têm uma diferença significativa entre as probabilidades. Essas são suas palavras-chave, tendendo a expressar um sentimento ou outro. Essas palavras têm muito peso na determinação dos sentimentos de seus tweets.

**Problemas de Probabilidade Zero:**
Agora, vamos dar uma olhada na palavra "porque". Como você pode ver, ela aparece apenas no corpus positivo. Sua probabilidade condicional para a classe negativa é 0. Quando isso acontece, você não tem como comparar entre os dois corpora, o que se tornará um problema para seus cálculos. Para evitar isso, você suavizará sua função de probabilidade.



**Aplicação da Regra de Inferência Naive Bayes:**
Vamos supor que você recebe um novo tweet de um amigo e o tweet diz: "Estou feliz hoje, estou aprendendo." Você quer usar a tabela de probabilidades para prever os sentimentos do tweet inteiro. Essa expressão é chamada de regra de inferência Naive Bayes para classificação binária. Essa expressão diz que você vai pegar o produto das probabilidades de cada palavra na classe positiva e dividir pela probabilidade na classe negativa.

Vamos calcular esse produto para este tweet. Para cada palavra, selecione suas probabilidades da tabela:
- Para "eu", você obtém uma probabilidade positiva de 0.2 e uma probabilidade negativa de 0.2. O que entra no produto é apenas $ \frac{0.2}{0.2} $.
- Para "estou", você também obtém $ \frac{0.2}{0.2} $.
- Para "feliz", você obtém $ \frac{0.14}{0.10} $.
- Para "hoje", você não encontra a palavra na tabela, significando que esta palavra não está em seu vocabulário. Então, você não incluirá nenhum termo na pontuação.
- Para a segunda ocorrência de "eu" novamente, você terá $ \frac{0.2}{0.2} $.
- Para a segunda ocorrência de "estou", você terá $ \frac{0.2}{0.2} $.
- "Aprendendo" tem $ \frac{0.10}{0.10} $.

Observe que todas as palavras neutras no tweet, como "eu" e "estou", simplesmente se cancelam na expressão. O que resta é $ \frac{0.14}{0.10} $, que é igual a 1.4. Esse valor é maior que um, o que significa que, no geral, as palavras no tweet têm maior probabilidade de corresponder a um sentimento positivo. Então, você conclui que o tweet é positivo.

# Laplacian Smoothing

Às vezes, ao tentar calcular a probabilidade de uma palavra aparecer depois de outra, você pode querer contar o número de vezes que essas duas palavras apareceram juntas, uma após a outra, dividido pelo número de vezes que a primeira palavra apareceu. Agora, o que acontece se as duas palavras nunca apareceram juntas no corpus de treinamento? Você obteria uma probabilidade zero, e a probabilidade de toda a sequência poderia cair para zero. Como podemos resolver esse problema? Vamos dar uma olhada e mergulhar na suavização de Laplace, uma técnica que você pode usar para evitar que suas probabilidades sejam zero.



A expressão usada para calcular a probabilidade condicional de uma palavra, dada a classe, é a frequência da palavra no corpus, mostrada aqui como freq(palavra | classe) dividida pelo número de palavras no corpus, ou Nclasse. Suavizar a função de probabilidade significa que você usará uma fórmula ligeiramente diferente da original. Note que eu adicionei um "1" no numerador. Esta pequena transformação evita que a probabilidade seja zero. No entanto, ela adiciona um novo termo a todas as frequências que não é corretamente normalizado por Nclasse. Para compensar isso, você adicionará um novo termo no denominador Vclasse, que é o número de palavras únicas no seu vocabulário, para contabilizar esse termo extra adicionado no numerador. Assim, todas as probabilidades em cada coluna somarão um. Este processo é chamado de suavização de Laplace.



Voltando a este exemplo, vamos usar a fórmula nele. A primeira coisa que você precisa calcular é o número de palavras únicas no seu vocabulário. Neste exemplo, você tem oito palavras únicas. Agora, vamos calcular a probabilidade para cada palavra na classe positiva. Para a palavra "eu" na classe positiva, você obtém (3+1)/(13+8), que é igual a 0,19. Para a classe negativa, você tem (3+1)/(12+8), que é igual a 0,2, e assim por diante para o resto da tabela. Os números mostrados aqui foram arredondados. Mas, usando esse método, a soma das probabilidades na sua tabela ainda será um. Note que a palavra "porque" não tem mais uma probabilidade de zero.

# Log Likelihood

#### Razões de Probabilidade

Para calcular a probabilidade condicional de uma palavra dado o sentimento (positivo ou negativo), você divide a frequência da palavra na classe pelo total de palavras na mesma classe. Vamos analisar como essas razões são importantes na classificação de sentimentos:



**Probabilidades Condicionais**:
   - Você calcula a razão de probabilidade para cada palavra no seu vocabulário. Por exemplo, a razão para a palavra "eu" é 0.2/0.2, que resulta em 1.
   - A razão para a palavra "feliz" é 0.14/0.1, que resulta em 1.4.
   - Palavras como "triste" e "não" têm razões menores que 1, indicando uma tendência negativa.

**Interpretação das Razões**:
   - Palavras neutras têm uma razão aproximadamente igual a 1.
   - Palavras positivas têm uma razão maior que 1; quanto maior a razão, mais positiva é a palavra.
   - Palavras negativas têm uma razão menor que 1; quanto menor a razão, mais negativa é a palavra.

**Aplicação na Classificação de Sentimentos**:
   - Para classificar um tweet, você calcula o produto das razões das palavras presentes no tweet.
   - Se o produto for maior que 1, o tweet é classificado como positivo. Se for menor que 1, é classificado como negativo.
   - Este produto é chamado de "verossimilhança" (likelihood).

#### Razão Anterior (Prior Ratio)

Quando os conjuntos de dados são balanceados (mesmo número de tweets positivos e negativos), a razão anterior (prior ratio) é 1 e pode ser desconsiderada. Porém, para conjuntos de dados desbalanceados, essa razão se torna importante para ajustar a probabilidade geral.


#### Uso de Logaritmos

#### Problema do Subfluxo Numérico

Na implementação do Naive Bayes para a classificação de sentimentos, você frequentemente precisa multiplicar muitas probabilidades com valores entre 0 e 1. Esse processo pode resultar em subfluxo numérico, onde os números se tornam tão pequenos que não podem ser armazenados corretamente no dispositivo. Para solucionar esse problema, utilizamos uma propriedade dos logaritmos.

#### Uso de Logaritmos para Evitar Subfluxo

A fórmula do Naive Bayes para calcular a pontuação envolve a multiplicação da razão anterior (prior) pela verossimilhança (likelihood). Em vez de trabalhar com o produto dessas probabilidades, podemos utilizar logaritmos para transformar a multiplicação em soma, o que é numericamente mais estável.

**Transformação Logarítmica**:
   - A fórmula original do Naive Bayes é: $$ \text{score} = \text{prior} \times \text{likelihood} $$
   - Ao aplicar logaritmos, a expressão se torna: $$ \log(\text{score}) = \log(\text{prior}) + \log(\text{likelihood}) $$

**Cálculo do Logaritmo das Razões**:
   - Para calcular o logaritmo da pontuação, precisamos calcular o logaritmo da razão de probabilidade de cada palavra.
   - Essa razão, chamada de Lambda, é calculada como: $$ \Lambda = \log \left( \frac{P(\text{word}|\text{positive})}{P(\text{word}|\text{negative})} \right) $$

   

**Exemplo Prático**:
   - Vamos calcular Lambda para as palavras no tweet "I'm happy because I'm learning".
   - Para a palavra "I", a razão é $ \frac{0.05}{0.05} $, e o logaritmo de 1 é 0, indicando neutralidade.
   - Para "am", a razão é $ \frac{0.04}{0.04} $, novamente resultando em um logaritmo de 0.
   - Para "happy", a razão é $ \frac{0.14}{0.1} $, e o logaritmo dessa razão é positivo, indicando um sentimento positivo.

#### Classificação com Logaritmos

**Soma dos Logaritmos**:
   - Em vez de multiplicar as probabilidades, somamos os logaritmos das razões (Lambda) para todas as palavras no tweet.
   - Se a soma dos logaritmos for maior que 0, o tweet é classificado como positivo. Se for menor que 0, é classificado como negativo.

### Como Fazer Inferência com Naive Bayes Utilizando Logaritmos

#### Inferência de Sentimento

Agora que você já entendeu como calcular a verossimilhança logarítmica (log likelihood), vamos ver como aplicar isso na prática para determinar o sentimento de um tweet. O processo é simples, especialmente porque a maior parte do trabalho já foi feita.

**Cálculo da Verossimilhança Logarítmica**:
   - A verossimilhança logarítmica de um tweet é calculada somando os valores de lambda de cada palavra no tweet.
   - Para cada palavra no tweet, adicione o valor correspondente de lambda. Se a palavra não estiver presente no dicionário de lambdas, seu valor é 0.



**Exemplo Prático**:
   - Considere o tweet "I'm happy because I'm learning".
   - Calcule a soma dos valores de lambda para cada palavra:
     - "I" : lambda = 0
     - "am" : lambda = 0
     - "happy" : lambda = 2.2
     - "because" : lambda = 0
     - "I'm" : lambda = 0
     - "learning" : lambda = 1.1
   - Soma dos lambdas: 0 + 0 + 2.2 + 0 + 0 + 1.1 = 3.3

**Interpretação do Resultado**:
   - Se a soma dos lambdas for maior que 0, o tweet é considerado positivo.
   - Se a soma dos lambdas for menor que 0, o tweet é considerado negativo.
   - No exemplo acima, a soma foi 3.3, indicando que o tweet é positivo.

#### Recapitulando

**Uso das Palavras de Força**:
   - No exemplo, as palavras "happy" e "learning" são palavras de força que carregam um sentimento positivo, influenciando diretamente o resultado.
   - As palavras neutras como "I" e "am" não contribuíram para o valor final.

**Log Likelihood e Limite de Decisão**:
   - O log likelihood é a soma dos valores de lambda para cada palavra no tweet.
   - O limite de decisão é 0: valores positivos indicam tweets positivos e valores negativos indicam tweets negativos.

# Training Naïve Bayes

### Como Treinar um Classificador Naive Bayes

Diferente da regressão logística ou aprendizado profundo, aqui não utilizamos o gradiente descendente. Em vez disso, contamos as frequências das palavras no corpus. Vamos criar passo a passo um modelo Naive Bayes para análise de sentimentos utilizando um corpus de tweets.

#### Passos para Treinamento do Modelo Naive Bayes

**Coleta de Dados**:
   - O primeiro passo em qualquer projeto de aprendizado supervisionado é reunir os dados para treinar e testar seu modelo.
   - Para análise de sentimentos de tweets, isso envolve obter um corpus de tweets e dividi-los em dois grupos: tweets positivos e tweets negativos.

**Pré-processamento**:
   - Esta etapa é fundamental para o sucesso do seu modelo. Consiste em cinco sub-etapas:
     **Converter textos para minúsculas**: Normaliza o texto.
     **Remover pontuação, URLs e handles**: Limpa o texto de elementos irrelevantes.
     **Remover stop words**: Elimina palavras comuns que não carregam muito significado.
     **Stemming**: Reduz as palavras às suas raízes comuns.
     **Tokenização**: Divide o documento em palavras ou tokens únicos.
   - Implementar esta pipeline de processamento será relativamente simples para as tarefas desta semana, mas no mundo real, esta etapa pode consumir a maior parte do tempo do projeto.

**Computação do Vocabulário e Frequências**:
   - Com um corpus limpo de tweets processados, você começa a computar o vocabulário para cada palavra em cada classe.
   - Este processo produzirá uma tabela como a mostrada anteriormente, onde você também computa a soma das palavras em cada classe no mesmo passo.

**Cálculo das Probabilidades Condicionais**:
   - A partir da tabela de frequências, você obtém a probabilidade condicional (probabilidade da palavra dado a classe) usando a fórmula de suavização de Laplace.
   - Esta tabela conterá valores maiores que 0 e refletem as probabilidades condicionais para cada palavra em cada classe.

**Cálculo do Fator Lambda**:
   - Calcule a pontuação lambda para cada palavra, que é o logaritmo da razão das probabilidades condicionais.

**Estimativa do Log Prior**:
   - Conte o número de tweets positivos e negativos.
   - O log prior é o logaritmo da razão do número de tweets positivos sobre o número de tweets negativos. Em conjuntos de dados balanceados, o log prior é igual a 0. Em conjuntos de dados desbalanceados, este termo se torna importante.

## Calcular as verossimilhanças para cada tweet

Para cada tweet, calculamos a verossimilhança de o tweet ser positivo e a verossimilhança de ser negativo. Calculamos em diferentes colunas o numerador e o denominador da razão de verossimilhança introduzida anteriormente.

$$\log \frac{P(\text{tweet}|\text{pos})}{P(\text{tweet}|\text{neg})} = \log(P(\text{tweet}|\text{pos})) - \log(P(\text{tweet}|\text{neg}))$$
$$\text{positivo} = \log(P(\text{tweet}|\text{pos})) = \sum_{i=0}^{n}{\log P(W_i|\text{pos})}$$
$$\text{negativo} = \log(P(\text{tweet}|\text{neg})) = \sum_{i=0}^{n}{\log P(W_i|\text{neg})}$$

# Usando Elipses de Confiança para interpretar Naïve Bayes

Nesta seção, usaremos a [elipse de confiança](https://matplotlib.org/3.1.1/gallery/statistics/confidence_ellipse.html#sphx-glr-gallery-statistics-confidence-ellipse-py) para nos dar uma ideia do que o modelo Naïve Bayes vê.

Uma elipse de confiança é uma maneira de visualizar uma variável aleatória em 2D. É uma forma melhor do que plotar os pontos em um plano cartesiano porque, com grandes conjuntos de dados, os pontos podem se sobrepor e esconder a verdadeira distribuição dos dados. As elipses de confiança resumem a informação do conjunto de dados com apenas quatro parâmetros:

* Centro: É a média numérica dos atributos.
* Altura e largura: Relacionadas com a variância de cada atributo. O usuário deve especificar a quantidade desejada de desvios padrão usados para plotar a elipse.
* Ângulo: Relacionado com a covariância entre os atributos.

O parâmetro __n_std__ representa o número de desvios padrão delimitados pela elipse. Lembre-se de que, para distribuições normais aleatórias:

* Cerca de 68% da área sob a curva está dentro de 1 desvio padrão em torno da média.
* Cerca de 95% da área sob a curva está dentro de 2 desvios padrão em torno da média.
* Cerca de 99,7% da área sob a curva está dentro de 3 desvios padrão em torno da média.



No próximo gráfico, plotaremos os dados e suas respectivas elipses de confiança usando 2 desvios padrão e 3 desvios padrão.

````python
from matplotlib.patches import Ellipse
import matplotlib.transforms as transforms

def confidence_ellipse(x, y, ax, n_std=3.0, facecolor='none', **kwargs):
    """
    Cria um gráfico da elipse de confiança de covariância de `x` e `y`

    Parâmetros
    ----------
    x, y : array_like, shape (n, )
        Dados de entrada.

    ax : matplotlib.axes.Axes
        O objeto de eixos para desenhar a elipse.

    n_std : float
        O número de desvios padrão para determinar os raios da elipse.

    Retornos
    -------
    matplotlib.patches.Ellipse

    Outros parâmetros
    -----------------
    kwargs : propriedades de `~matplotlib.patches.Patch`
    """
    if x.size != y.size:
        raise ValueError("x e y devem ter o mesmo tamanho")

    cov = np.cov(x, y)
    pearson = cov[0, 1] / np.sqrt(cov[0, 0] * cov[1, 1])
    # Usando um caso especial para obter os autovalores deste
    # conjunto de dados bidimensional.
    ell_radius_x = np.sqrt(1 + pearson)
    ell_radius_y = np.sqrt(1 - pearson)
    ellipse = Ellipse((0, 0),
                      width=ell_radius_x * 2,
                      height=ell_radius_y * 2,
                      facecolor=facecolor,
                      **kwargs)

    # Calculando o desvio padrão de x a partir
    # da raiz quadrada da variância e multiplicando
    # pelo número dado de desvios padrão.
    scale_x = np.sqrt(cov[0, 0]) * n_std
    mean_x = np.mean(x)

    # Calculando o desvio padrão de y ...
    scale_y = np.sqrt(cov[1, 1]) * n_std
    mean_y = np.mean(y)

    transf = transforms.Affine2D() \
        .rotate_deg(45) \
        .scale(scale_x, scale_y) \
        .translate(mean_x, mean_y)

    ellipse.set_transform(transf + ax.transData)
    return ax.add_patch(ellipse)
````

# Naïve Bayes Assumptions

Agora vou falar sobre as suposições subjacentes ao método Naive Bayes. A principal delas é a independência das palavras em uma sentença. E vou te mostrar por que isso pode ser um grande problema quando o método é aplicado. Naive Bayes é um modelo muito simples porque não requer a definição de parâmetros personalizados. Esse método é chamado de "naive" devido às suposições que faz sobre os dados. A primeira suposição é a independência entre os preditores ou características associadas a cada classe. A segunda está relacionada ao seu conjunto de validação. Vamos explorar cada uma dessas suposições e como elas podem afetar seus resultados.



Para ilustrar o que a independência entre características significa, vamos olhar para a seguinte frase: "Está ensolarado e quente no deserto do Saara." Naive Bayes assume que as palavras em um texto são independentes umas das outras. Mas, como você pode ver, isso geralmente não é o caso. As palavras "ensolarado" e "quente" frequentemente aparecem juntas, como neste exemplo. Juntas, elas podem estar relacionadas à coisa que estão descrevendo, como uma praia ou um deserto. Portanto, as palavras na frase não são sempre necessariamente independentes umas das outras. Mas o Naive Bayes assume que são, o que pode levar você a subestimar ou superestimar as probabilidades condicionais de palavras individuais.



Por exemplo, ao usar o Naive Bayes, se sua tarefa fosse completar a frase "Está sempre frio e nevado em __.", o Naive Bayes poderia atribuir probabilidade igual às palavras primavera, verão, outono e inverno. Embora pelo contexto você possa ver que inverno deve ser o candidato mais provável.

Outro problema com o Naive Bayes é que ele depende da distribuição dos conjuntos de dados de treinamento. Um bom conjunto de dados conterá a mesma proporção de tweets positivos e negativos que uma amostra aleatória. No entanto, a maioria dos corpos anotados disponíveis são artificialmente balanceados, assim como o conjunto de dados que você usa para a tarefa. Na verdade, tweets positivos tendem a ocorrer com mais frequência do que seus equivalentes negativos. Uma razão para isso é que tweets negativos podem conter conteúdo banido pela plataforma ou silenciado pelo usuário, como vocabulário impróprio ou ofensivo. Assumindo que a realidade se comporta como seu corpus de treinamento, isso pode resultar em um modelo muito otimista ou muito pessimista.

## Processar os Dados

Para qualquer projeto de aprendizado de máquina, uma vez que você tenha coletado os dados, o primeiro passo é processá-los para criar entradas úteis para o seu modelo.
- **Remover ruído**: Primeiro, você vai querer remover o ruído dos seus dados -- ou seja, remover palavras que não dizem muito sobre o conteúdo. Isso inclui todas as palavras comuns como 'eu, você, são, é, etc...' que não nos dariam informações suficientes sobre o sentimento.
- Também vamos remover tickers de mercado de ações, símbolos de retweet, hyperlinks e hashtags porque eles não nos dizem muito sobre o sentimento.
- Você também quer remover toda a pontuação de um tweet. A razão para fazer isso é que queremos tratar palavras com ou sem pontuação como a mesma palavra, em vez de tratar "feliz", "feliz?", "feliz!", "feliz," e "feliz." como palavras diferentes.
- Finalmente, você quer usar stemming para manter apenas uma variação de cada palavra. Em outras palavras, vamos tratar "motivação", "motivado" e "motivar" de maneira similar, agrupando-as dentro do mesmo radical "motiv-".

Nós fornecemos a função `process_tweet` que faz isso para você.


````python
import re                                  # Biblioteca para operações de expressões regulares

from nltk.corpus import stopwords          # Módulo para stopwords que vem com NLTK
from nltk.stem import PorterStemmer        # Módulo para stemming
from nltk.tokenize import TweetTokenizer   # Módulo para tokenização de strings
````

````python
def process_tweet(tweet):
    """Função de processamento de tweet.
    Entrada:
        tweet: uma string contendo um tweet
    Saída:
        tweets_clean: uma lista de palavras contendo o tweet processado
    """
    stemmer = PorterStemmer()
    stopwords_english = stopwords.words('english')
    
    # Remove tickers do mercado de ações, como $GE
    tweet = re.sub(r'\$\w*', '', tweet)
    
    # Remove texto de retweet no estilo antigo "RT"
    tweet = re.sub(r'^RT[\s]+', '', tweet)
    
    # Remove hyperlinks    
    tweet = re.sub(r'https?://[^\s\n\r]+', '', tweet)
    
    # Remove hashtags, apenas o símbolo # da palavra
    tweet = re.sub(r'#', '', tweet)
    
    # Tokeniza o tweet
    tokenizer = TweetTokenizer(preserve_case=False, strip_handles=True, reduce_len=True)
    tweet_tokens = tokenizer.tokenize(tweet)

    tweets_clean = []
    for word in tweet_tokens:
        if (word not in stopwords_english and  # Remove stopwords
                word not in string.punctuation):  # Remove pontuação
            stem_word = stemmer.stem(word)  # Aplica stemming na palavra
            tweets_clean.append(stem_word)

    return tweets_clean
````

### count_tweets
- A chave no dicionário é uma tupla contendo a palavra radicalizada e seu rótulo, por exemplo, ("feliz", 1).
- O valor é o número de vezes que essa palavra aparece na coleção de tweets dada (um inteiro).

````python
def count_tweets(result, tweets, ys):
    '''
    Input:
        result: um dicionário que será usado para mapear cada par à sua frequência
        tweets: uma lista de tweets
        ys: uma lista correspondente ao sentimento de cada tweet (0 ou 1)
    Output:
        result: um dicionário mapeando cada par à sua frequência
    '''
  
    for y, tweet in zip(ys, tweets):
        for word in process_tweet(tweet):
            # define a chave, que é a tupla palavra e rótulo
            pair = (word, y)
            
            # se a chave existir no dicionário, incrementa a contagem
            if pair in result:
                result[pair] += 1

            # caso contrário, se a chave for nova, adiciona ao dicionário e define a contagem como 1
            else:
                result[pair] = 1

    return result
````

## Treine seu Modelo usando Naive Bayes

Naive Bayes é um algoritmo que pode ser usado para análise de sentimento. Ele leva pouco tempo para treinar e também tem um tempo de previsão curto.

#### Então, como você treina um classificador de Naive Bayes?
- A primeira parte do treinamento de um classificador de Naive Bayes é identificar o número de classes que você tem.
- Você criará uma probabilidade para cada classe.
$P(D_{pos})$ é a probabilidade de que o documento seja positivo.
$P(D_{neg})$ é a probabilidade de que o documento seja negativo.
Use as fórmulas a seguir e armazene os valores em um dicionário:

$$P(D_{pos}) = \frac{D_{pos}}{D}\tag{1}$$

$$P(D_{neg}) = \frac{D_{neg}}{D}\tag{2}$$

Onde $D$ é o número total de documentos, ou tweets neste caso, $D_{pos}$ é o número total de tweets positivos e $D_{neg}$ é o número total de tweets negativos.

#### Prior e Logprior

A probabilidade a priori representa a probabilidade subjacente na população-alvo de que um tweet seja positivo versus negativo. Em outras palavras, se não tivéssemos nenhuma informação específica e escolhêssemos cegamente um tweet do conjunto populacional, qual seria a probabilidade de ele ser positivo versus negativo? Isso é o "prior".

O prior é a razão das probabilidades $\frac{P(D_{pos})}{P(D_{neg})}$. Podemos pegar o log do prior para reescalá-lo, e chamaremos isso de logprior.

$\text{logprior} = \log \left( \frac{P(D_{pos})}{P(D_{neg})} \right) = \log \left( \frac{D_{pos}}{D_{neg}} \right)$.

Observe que $\log(\frac{A}{B})$ é o mesmo que $\log(A) - \log(B)$. Então, o logprior também pode ser calculado como a diferença entre dois logs:

$$\text{logprior} = \log (P(D_{pos})) - \log (P(D_{neg})) = \log (D_{pos}) - \log (D_{neg})\tag{3}$$


#### Probabilidade Positiva e Negativa de uma Palavra
Para calcular a probabilidade positiva e a probabilidade negativa para uma palavra específica no vocabulário, usaremos as seguintes entradas:

- $freq_{pos}$ e $freq_{neg}$ são as frequências dessa palavra específica na classe positiva ou negativa. Em outras palavras, a frequência positiva de uma palavra é o número de vezes que a palavra é contada com o rótulo 1.
- $N_{pos}$ e $N_{neg}$ são o número total de palavras positivas e negativas para todos os documentos (para todos os tweets), respectivamente.
- $V$ é o número de palavras únicas em todo o conjunto de documentos, para todas as classes, sejam elas positivas ou negativas.

Usaremos esses valores para calcular a probabilidade positiva e a probabilidade negativa para uma palavra específica usando esta fórmula:

$$ P(W_{pos}) = \frac{freq_{pos} + 1}{N_{pos} + V}\tag{4} $$
$$ P(W_{neg}) = \frac{freq_{neg} + 1}{N_{neg} + V}\tag{5} $$

Observe que adicionamos "+1" no numerador para suavização aditiva. Este [artigo da wiki](https://en.wikipedia.org/wiki/Additive_smoothing) explica mais sobre suavização aditiva.

#### Log likelihood
Para calcular a logverossimilhança dessa mesma palavra, podemos implementar as seguintes equações:

$$\text{loglikelihood} = \log \left(\frac{P(W_{pos})}{P(W_{neg})} \right)\tag{6}$$

````python
# Dicionário para armazenar a contagem de palavras
freqs = count_tweets({}, train_x, train_y)
````

````python
def train_naive_bayes(freqs, train_x, train_y):
    '''
    Input:
        freqs: dicionário de (palavra, rótulo) para a frequência da palavra
        train_x: uma lista de tweets
        train_y: uma lista de rótulos correspondentes aos tweets (0,1)
    Output:
        logprior: o logaritmo da probabilidade a priori
        loglikelihood: o logaritmo da verossimilhança da equação de Naive Bayes
    '''
    loglikelihood = {}
    logprior = 0

    # calcular V, o número de palavras únicas no vocabulário
    vocab = set([pair[0] for pair in freqs.keys()])
    V = len(vocab)

    # calcular N_pos, N_neg
    N_pos = sum([freqs.get((word, 1), 0) for word in vocab])
    N_neg = sum([freqs.get((word, 0), 0) for word in vocab])

    # Calcular D, o número de documentos
    D = len(train_x)

    # Calcular D_pos, o número de documentos positivos
    D_pos = sum(train_y)

    # Calcular D_neg, o número de documentos negativos
    D_neg = D - D_pos

    # Calcular logprior
    logprior = np.log(D_pos) - np.log(D_neg)

    # Para cada palavra no vocabulário...
    for word in vocab:
        # obter a frequência positiva e negativa da palavra
        freq_pos = freqs.get((word, 1), 0)
        freq_neg = freqs.get((word, 0), 0)

        # calcular a probabilidade de que cada palavra seja positiva e negativa
        p_w_pos = (freq_pos + 1) / (N_pos + V)
        p_w_neg = (freq_neg + 1) / (N_neg + V)

        # calcular o logaritmo da verossimilhança da palavra
        loglikelihood[word] = np.log(p_w_pos / p_w_neg)

    return logprior, loglikelihood
````

````python
logprior, loglikelihood = train_naive_bayes(freqs, train_x, train_y)
print(logprior)
print(len(loglikelihood))
````

## Teste seu Naive Bayes

Agora que temos o logprior e a loglikelihood, podemos testar a função Naive Bayes fazendo previsões em alguns tweets!

### naive_bayes_predict

**Instruções**:

* A função recebe um tweet, logprior, e loglikelihood.
* Ela retorna a probabilidade de que o tweet pertença à classe positiva ou negativa.
* Para cada tweet, some as log-verossimilhanças de cada palavra no tweet.
* Adicione também o logprior a essa soma para obter o sentimento previsto desse tweet.

$$ p = \text{logprior} + \sum_i^N (\text{loglikelihood}_i)$$

#### Nota
Observe que calculamos o prior a partir dos dados de treinamento, e que os dados de treinamento estão igualmente divididos entre rótulos positivos e negativos (4000 tweets positivos e 4000 tweets negativos). Isso significa que a razão de positivo para negativo é 1, e o logprior é 0.

O valor de 0.0 significa que, quando adicionamos o logprior à loglikelihood, estamos apenas adicionando zero à loglikelihood. No entanto, lembre-se de incluir o logprior, porque sempre que os dados não estiverem perfeitamente balanceados, o logprior será um valor diferente de zero.

````python
def naive_bayes_predict(tweet, logprior, loglikelihood):
    '''
    Input:
        tweet: uma string
        logprior: um número
        loglikelihood: um dicionário de palavras mapeando para números
    Output:
        p: a soma de todas as logverossimilhanças de cada palavra no tweet (se encontradas no dicionário) + logprior (um número)

    '''
  
    # processa o tweet para obter uma lista de palavras
    word_l = process_tweet(tweet)

    # inicializa a probabilidade para zero
    p = 0

    # adiciona o logprior
    p += logprior

    for word in word_l:

        # verifica se a palavra existe no dicionário de logverossimilhanças
        if word in loglikelihood:
            # adiciona a logverossimilhança dessa palavra à probabilidade
            p += loglikelihood[word]


    return p
````

### test_naive_bayes
Implemente test_naive_bayes.

**Instruções**:
* Implemente `test_naive_bayes` para verificar a precisão de suas previsões.
* A função recebe `test_x`, `test_y`, log_prior, e loglikelihood.
* Ela retorna a precisão do seu modelo.
* Primeiro, use a função `naive_bayes_predict` para fazer previsões para cada tweet em test_x.

````python
def test_naive_bayes(test_x, test_y, logprior, loglikelihood, naive_bayes_predict=naive_bayes_predict):
    """
    Input:
        test_x: Uma lista de tweets
        test_y: os rótulos correspondentes para a lista de tweets
        logprior: o logprior
        loglikelihood: um dicionário com as logverossimilhanças para cada palavra
    Output:
        accuracy: (# de tweets classificados corretamente)/(total # de tweets)
    """
    accuracy = 0  # retorne isso corretamente

    # lista para armazenar os valores previstos
    y_hats = [naive_bayes_predict(tweet, logprior, loglikelihood) for tweet in test_x]
    
    # 0 se o valor previsto for negativo e 1 se o valor previsto for positivo
    y_hats_binary = [1 if y_hat > 0 else 0 for y_hat in y_hats]

    # calcular a quantidade de valores previstos corretamente
    correct_predictions = np.sum(np.array(y_hats_binary) == np.array(test_y))

    # calcular a precisão
    accuracy = correct_predictions / len(test_y)

    return accuracy
````

````python
print("Naive Bayes accuracy = %0.4f" %
      (test_naive_bayes(test_x, test_y, logprior, loglikelihood)))
````

## Filtrar palavras por Razão de Contagens Positivas para Negativas

- Algumas palavras têm mais contagens positivas do que outras e podem ser consideradas "mais positivas". Da mesma forma, algumas palavras podem ser consideradas mais negativas do que outras.
- Uma maneira de definir o nível de positividade ou negatividade, sem calcular a verossimilhança, é comparar a frequência positiva com a frequência negativa da palavra.
    - Observe que também podemos usar os cálculos de verossimilhança para comparar a positividade ou negatividade relativa das palavras.
- Podemos calcular a razão de frequências positivas para negativas de uma palavra.
- Assim que conseguirmos calcular essas razões, também podemos filtrar um subconjunto de palavras que tenham uma razão mínima de positividade/negatividade ou maior.
- Da mesma forma, também podemos filtrar um subconjunto de palavras que tenham uma razão máxima de positividade/negatividade ou menor (palavras que são pelo menos tão negativas ou até mais negativas do que um determinado limite).


### get_ratio

- Dado o dicionário freqs de palavras e uma palavra em particular, use `lookup(freqs,word,1)` para obter a contagem positiva da palavra.
- Da mesma forma, use a função `lookup` para obter a contagem negativa dessa palavra.
- Calcule a razão das contagens positivas divididas pelas contagens negativas

 $$ratio = \frac{\text{pos\_words} + 1}{\text{neg\_words} + 1}$$

Onde pos_words e neg_words correspondem à frequência das palavras em suas respectivas classes.

````python
def lookup(freqs, word, label):
    '''
    Entrada:
        freqs: um dicionário com a frequência de cada par (ou tupla)
        word: a palavra a ser procurada
        label: o rótulo correspondente à palavra
    Saída:
        n: o número de vezes que a palavra com seu rótulo correspondente aparece.
    '''
    n = 0  # freqs.get((word, label), 0)

    pair = (word, label)
    if pair in freqs:
        n = freqs[pair]

    return n
````

````python
def get_ratio(freqs, word):
    '''
    Entrada:
        freqs: dicionário contendo as palavras

    Saída: um dicionário com as chaves 'positive', 'negative' e 'ratio'.
        Exemplo: {'positive': 10, 'negative': 20, 'ratio': 0.5}
    '''
    pos_neg_ratio = {'positive': 0, 'negative': 0, 'ratio': 0.0}
    
    # use lookup() para encontrar as contagens positivas para a palavra (denotada pelo inteiro 1)
    pos_neg_ratio['positive'] = lookup(freqs, word, 1)
    
    # use lookup() para encontrar as contagens negativas para a palavra (denotada pelo inteiro 0)
    pos_neg_ratio['negative'] = lookup(freqs, word, 0)
    
    # calcular a razão de contagens positivas para negativas para a palavra
    pos_neg_ratio['ratio'] = (pos_neg_ratio['positive'] + 1) / (pos_neg_ratio['negative'] + 1)
  
    return pos_neg_ratio
````

### get_words_by_threshold

* Se definirmos o rótulo como 1, procuraremos todas as palavras cujo limiar de positivo/negativo seja pelo menos tão alto quanto esse limiar ou maior.
* Se definirmos o rótulo como 0, procuraremos todas as palavras cujo limiar de positivo/negativo seja no máximo tão baixo quanto o limite dado ou menor.
* Use a função `get_ratio` para obter um dicionário contendo a contagem positiva, contagem negativa e a razão das contagens positivas para negativas.
* Anexe o dicionário `get_ratio` dentro de outro dicionário, onde a chave é a palavra e o valor é o dicionário `pos_neg_ratio` que é retornado pela função `get_ratio`.
Um par chave-valor de exemplo teria esta estrutura:
```
{'happi':
    {'positive': 10, 'negative': 20, 'ratio': 0.524}
}
```
```

````python
def get_words_by_threshold(freqs, label, threshold, get_ratio=get_ratio):
    '''
    Entrada:
        freqs: dicionário de palavras
        label: 1 para positivo, 0 para negativo
        threshold: razão que será usada como limite para incluir uma palavra no dicionário retornado
    Saída:
        word_list: dicionário contendo a palavra e informações sobre sua contagem positiva, contagem negativa e razão de contagens positivas para negativas.
        exemplo de um par chave-valor:
        {'happi':
            {'positive': 10, 'negative': 20, 'ratio': 0.5}
        }
    '''
    word_list = {}
    for key in freqs.keys():
        word, _ = key

        # obter a razão positiva/negativa para uma palavra
        pos_neg_ratio = get_ratio(freqs, word)

        # se o rótulo for 1 e a razão for maior ou igual ao limite...
        if label == 1 and pos_neg_ratio['ratio'] >= threshold:
        
            # Adicionar a razão positiva/negativa ao dicionário
            word_list[word] = pos_neg_ratio

        # Se o rótulo for 0 e a razão positiva/negativa for menor ou igual ao limite...
        elif label == 0 and pos_neg_ratio['ratio'] <= threshold:
        
            # Adicionar a razão positiva/negativa ao dicionário
            word_list[word] = pos_neg_ratio

        # caso contrário, não incluir essa palavra na lista (não fazer nada)

    return word_list
````
