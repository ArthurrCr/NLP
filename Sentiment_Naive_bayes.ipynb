# Log Likelihood

#### RazÃµes de Probabilidade

Para calcular a probabilidade condicional de uma palavra dado o sentimento (positivo ou negativo), vocÃª divide a frequÃªncia da palavra na classe pelo total de palavras na mesma classe. Vamos analisar como essas razÃµes sÃ£o importantes na classificaÃ§Ã£o de sentimentos:



**Probabilidades Condicionais**:
   - VocÃª calcula a razÃ£o de probabilidade para cada palavra no seu vocabulÃ¡rio. Por exemplo, a razÃ£o para a palavra "eu" Ã© 0.2/0.2, que resulta em 1.
   - A razÃ£o para a palavra "feliz" Ã© 0.14/0.1, que resulta em 1.4.
   - Palavras como "triste" e "nÃ£o" tÃªm razÃµes menores que 1, indicando uma tendÃªncia negativa.

**InterpretaÃ§Ã£o das RazÃµes**:
   - Palavras neutras tÃªm uma razÃ£o aproximadamente igual a 1.
   - Palavras positivas tÃªm uma razÃ£o maior que 1; quanto maior a razÃ£o, mais positiva Ã© a palavra.
   - Palavras negativas tÃªm uma razÃ£o menor que 1; quanto menor a razÃ£o, mais negativa Ã© a palavra.

**AplicaÃ§Ã£o na ClassificaÃ§Ã£o de Sentimentos**:
   - Para classificar um tweet, vocÃª calcula o produto das razÃµes das palavras presentes no tweet.
   - Se o produto for maior que 1, o tweet Ã© classificado como positivo. Se for menor que 1, Ã© classificado como negativo.
   - Este produto Ã© chamado de "verossimilhanÃ§a" (likelihood).

#### RazÃ£o Anterior (Prior Ratio)

Quando os conjuntos de dados sÃ£o balanceados (mesmo nÃºmero de tweets positivos e negativos), a razÃ£o anterior (prior ratio) Ã© 1 e pode ser desconsiderada. PorÃ©m, para conjuntos de dados desbalanceados, essa razÃ£o se torna importante para ajustar a probabilidade geral.


#### Uso de Logaritmos

#### Problema do Subfluxo NumÃ©rico

Na implementaÃ§Ã£o do Naive Bayes para a classificaÃ§Ã£o de sentimentos, vocÃª frequentemente precisa multiplicar muitas probabilidades com valores entre 0 e 1. Esse processo pode resultar em subfluxo numÃ©rico, onde os nÃºmeros se tornam tÃ£o pequenos que nÃ£o podem ser armazenados corretamente no dispositivo. Para solucionar esse problema, utilizamos uma propriedade dos logaritmos.

#### Uso de Logaritmos para Evitar Subfluxo

A fÃ³rmula do Naive Bayes para calcular a pontuaÃ§Ã£o envolve a multiplicaÃ§Ã£o da razÃ£o anterior (prior) pela verossimilhanÃ§a (likelihood). Em vez de trabalhar com o produto dessas probabilidades, podemos utilizar logaritmos para transformar a multiplicaÃ§Ã£o em soma, o que Ã© numericamente mais estÃ¡vel.

**TransformaÃ§Ã£o LogarÃ­tmica**:
   - A fÃ³rmula original do Naive Bayes Ã©: $$ \text{score} = \text{prior} \times \text{likelihood} $$
   - Ao aplicar logaritmos, a expressÃ£o se torna: $$ \log(\text{score}) = \log(\text{prior}) + \log(\text{likelihood}) $$

**CÃ¡lculo do Logaritmo das RazÃµes**:
   - Para calcular o logaritmo da pontuaÃ§Ã£o, precisamos calcular o logaritmo da razÃ£o de probabilidade de cada palavra.
   - Essa razÃ£o, chamada de Lambda, Ã© calculada como: $$ \Lambda = \log \left( \frac{P(\text{word}|\text{positive})}{P(\text{word}|\text{negative})} \right) $$

   

**Exemplo PrÃ¡tico**:
   - Vamos calcular Lambda para as palavras no tweet "I'm happy because I'm learning".
   - Para a palavra "I", a razÃ£o Ã© $ \frac{0.05}{0.05} $, e o logaritmo de 1 Ã© 0, indicando neutralidade.
   - Para "am", a razÃ£o Ã© $ \frac{0.04}{0.04} $, novamente resultando em um logaritmo de 0.
   - Para "happy", a razÃ£o Ã© $ \frac{0.14}{0.1} $, e o logaritmo dessa razÃ£o Ã© positivo, indicando um sentimento positivo.

#### ClassificaÃ§Ã£o com Logaritmos

**Soma dos Logaritmos**:
   - Em vez de multiplicar as probabilidades, somamos os logaritmos das razÃµes (Lambda) para todas as palavras no tweet.
   - Se a soma dos logaritmos for maior que 0, o tweet Ã© classificado como positivo. Se for menor que 0, Ã© classificado como negativo.

## Calcular as verossimilhanÃ§as para cada tweet

Para cada tweet, calculamos a verossimilhanÃ§a de o tweet ser positivo e a verossimilhanÃ§a de ser negativo. Calculamos em diferentes colunas o numerador e o denominador da razÃ£o de verossimilhanÃ§a introduzida anteriormente.

$$\log \frac{P(\text{tweet}|\text{pos})}{P(\text{tweet}|\text{neg})} = \log(P(\text{tweet}|\text{pos})) - \log(P(\text{tweet}|\text{neg}))$$

# Usando Elipses de ConfianÃ§a para interpretar NaÃ¯ve Bayes

Nesta seÃ§Ã£o, usaremos a [elipse de confianÃ§a](https://matplotlib.org/3.1.1/gallery/statistics/confidence_ellipse.html#sphx-glr-gallery-statistics-confidence-ellipse-py) para nos dar uma ideia do que o modelo NaÃ¯ve Bayes vÃª.

Uma elipse de confianÃ§a Ã© uma maneira de visualizar uma variÃ¡vel aleatÃ³ria em 2D. Ã‰ uma forma melhor do que plotar os pontos em um plano cartesiano porque, com grandes conjuntos de dados, os pontos podem se sobrepor e esconder a verdadeira distribuiÃ§Ã£o dos dados. As elipses de confianÃ§a resumem a informaÃ§Ã£o do conjunto de dados com apenas quatro parÃ¢metros:

* Centro: Ã‰ a mÃ©dia numÃ©rica dos atributos.
* Altura e largura: Relacionadas com a variÃ¢ncia de cada atributo. O usuÃ¡rio deve especificar a quantidade desejada de desvios padrÃ£o usados para plotar a elipse.
* Ã‚ngulo: Relacionado com a covariÃ¢ncia entre os atributos.

O parÃ¢metro __n_std__ representa o nÃºmero de desvios padrÃ£o delimitados pela elipse. Lembre-se de que, para distribuiÃ§Ãµes normais aleatÃ³rias:

* Cerca de 68% da Ã¡rea sob a curva estÃ¡ dentro de 1 desvio padrÃ£o em torno da mÃ©dia.
* Cerca de 95% da Ã¡rea sob a curva estÃ¡ dentro de 2 desvios padrÃ£o em torno da mÃ©dia.
* Cerca de 99,7% da Ã¡rea sob a curva estÃ¡ dentro de 3 desvios padrÃ£o em torno da mÃ©dia.



No prÃ³ximo grÃ¡fico, plotaremos os dados e suas respectivas elipses de confianÃ§a usando 2 desvios padrÃ£o e 3 desvios padrÃ£o.

````python
from matplotlib.patches import Ellipse
import matplotlib.transforms as transforms

def confidence_ellipse(x, y, ax, n_std=3.0, facecolor='none', **kwargs):
    """
    Cria um grÃ¡fico da elipse de confianÃ§a de covariÃ¢ncia de `x` e `y`

    ParÃ¢metros
    ----------
    x, y : array_like, shape (n, )
        Dados de entrada.

    ax : matplotlib.axes.Axes
        O objeto de eixos para desenhar a elipse.

    n_std : float
        O nÃºmero de desvios padrÃ£o para determinar os raios da elipse.

    Retornos
    -------
    matplotlib.patches.Ellipse

    Outros parÃ¢metros
    -----------------
    kwargs : propriedades de `~matplotlib.patches.Patch`
    """
    if x.size != y.size:
        raise ValueError("x e y devem ter o mesmo tamanho")

    cov = np.cov(x, y)
    pearson = cov[0, 1] / np.sqrt(cov[0, 0] * cov[1, 1])
    # Usando um caso especial para obter os autovalores deste
    # conjunto de dados bidimensional.
    ell_radius_x = np.sqrt(1 + pearson)
    ell_radius_y = np.sqrt(1 - pearson)
    ellipse = Ellipse((0, 0),
                      width=ell_radius_x * 2,
                      height=ell_radius_y * 2,
                      facecolor=facecolor,
                      **kwargs)

    # Calculando o desvio padrÃ£o de x a partir
    # da raiz quadrada da variÃ¢ncia e multiplicando
    # pelo nÃºmero dado de desvios padrÃ£o.
    scale_x = np.sqrt(cov[0, 0]) * n_std
    mean_x = np.mean(x)

    # Calculando o desvio padrÃ£o de y ...
    scale_y = np.sqrt(cov[1, 1]) * n_std
    mean_y = np.mean(y)

    transf = transforms.Affine2D() \
        .rotate_deg(45) \
        .scale(scale_x, scale_y) \
        .translate(mean_x, mean_y)

    ellipse.set_transform(transf + ax.transData)
    return ax.add_patch(ellipse)
````

````python
# Plote as amostras usando as colunas 1 e 2 da matriz
fig, ax = plt.subplots(figsize=(8, 8))

cores = ['red', 'green']  # Defina uma paleta de cores
sentimentos = ['negativo', 'positivo'] 
index = data.index

# Cor com base no sentimento
for sentimento in data.sentiment.unique():
    ix = index[data.sentiment == sentimento]
    ax.scatter(data.iloc[ix].positive, data.iloc[ix].negative, c=cores[int(sentimento)], s=0.1, marker='*', label=sentimentos[int(sentimento)])

# Limites personalizados para este grÃ¡fico
plt.xlim(-200, 40)  
plt.ylim(-200, 40)

plt.xlabel("Positivo")  # RÃ³tulo do eixo x
plt.ylabel("Negativo")  # RÃ³tulo do eixo y

data_pos = data[data.sentiment == 1]  # Filtre apenas as amostras positivas
data_neg = data[data.sentiment == 0]  # Filtre apenas as amostras negativas

# Imprima elipses de confianÃ§a de 2 desvios padrÃ£o
confidence_ellipse(data_pos.positive, data_pos.negative, ax, n_std=2, edgecolor='black', label=r'$2\sigma$')
confidence_ellipse(data_neg.positive, data_neg.negative, ax, n_std=2, edgecolor='orange')

# Imprima elipses de confianÃ§a de 3 desvios padrÃ£o
confidence_ellipse(data_pos.positive, data_pos.negative, ax, n_std=3, edgecolor='black', linestyle=':', label=r'$3\sigma$')
confidence_ellipse(data_neg.positive, data_neg.negative, ax, n_std=3, edgecolor='orange', linestyle=':')
ax.legend(loc='lower right')

plt.show()
````

````python
import re                                  # Biblioteca para operaÃ§Ãµes de expressÃµes regulares

from nltk.corpus import stopwords          # MÃ³dulo para stopwords que vem com NLTK
from nltk.stem import PorterStemmer        # MÃ³dulo para stemming
from nltk.tokenize import TweetTokenizer   # MÃ³dulo para tokenizaÃ§Ã£o de strings
````

````python
def process_tweet(tweet):
    """FunÃ§Ã£o de processamento de tweet.
    Entrada:
        tweet: uma string contendo um tweet
    SaÃ­da:
        tweets_clean: uma lista de palavras contendo o tweet processado
    """
    stemmer = PorterStemmer()
    stopwords_english = stopwords.words('english')
    
    # Remove tickers do mercado de aÃ§Ãµes, como $GE
    tweet = re.sub(r'\$\w*', '', tweet)
    
    # Remove texto de retweet no estilo antigo "RT"
    tweet = re.sub(r'^RT[\s]+', '', tweet)
    
    # Remove hyperlinks    
    tweet = re.sub(r'https?://[^\s\n\r]+', '', tweet)
    
    # Remove hashtags, apenas o sÃ­mbolo # da palavra
    tweet = re.sub(r'#', '', tweet)
    
    # Tokeniza o tweet
    tokenizer = TweetTokenizer(preserve_case=False, strip_handles=True, reduce_len=True)
    tweet_tokens = tokenizer.tokenize(tweet)

    tweets_clean = []
    for word in tweet_tokens:
        if (word not in stopwords_english and  # Remove stopwords
                word not in string.punctuation):  # Remove pontuaÃ§Ã£o
            stem_word = stemmer.stem(word)  # Aplica stemming na palavra
            tweets_clean.append(stem_word)

    return tweets_clean
````

### Implementando suas FunÃ§Ãµes Auxiliares

Para ajudar a treinar seu modelo de Naive Bayes, vocÃª precisarÃ¡ calcular um dicionÃ¡rio onde as chaves sÃ£o uma tupla (palavra, rÃ³tulo) e os valores sÃ£o a frequÃªncia correspondente. Observe que os rÃ³tulos que usaremos aqui sÃ£o 1 para positivo e 0 para negativo.

VocÃª tambÃ©m implementarÃ¡ uma funÃ§Ã£o auxiliar de busca que recebe o dicionÃ¡rio `freqs`, uma palavra e um rÃ³tulo (1 ou 0) e retorna o nÃºmero de vezes que essa tupla palavra e rÃ³tulo aparece na coleÃ§Ã£o de tweets.

Por exemplo: dada uma lista de tweets `["eu estou bastante animado", "vocÃª estÃ¡ bastante feliz"]` e o rÃ³tulo 1, a funÃ§Ã£o retornarÃ¡ um dicionÃ¡rio que contÃ©m os seguintes pares chave-valor:

}

### count_tweets
- A chave no dicionÃ¡rio Ã© uma tupla contendo a palavra radicalizada e seu rÃ³tulo, por exemplo, ("feliz", 1).
- O valor Ã© o nÃºmero de vezes que essa palavra aparece na coleÃ§Ã£o de tweets dada (um inteiro).

````python
def count_tweets(result, tweets, ys):
    '''
    Input:
        result: um dicionÃ¡rio que serÃ¡ usado para mapear cada par Ã  sua frequÃªncia
        tweets: uma lista de tweets
        ys: uma lista correspondente ao sentimento de cada tweet (0 ou 1)
    Output:
        result: um dicionÃ¡rio mapeando cada par Ã  sua frequÃªncia
    '''
  
    for y, tweet in zip(ys, tweets):
        for word in process_tweet(tweet):
            # define a chave, que Ã© a tupla palavra e rÃ³tulo
            pair = (word, y)
            
            # se a chave existir no dicionÃ¡rio, incrementa a contagem
            if pair in result:
                result[pair] += 1

            # caso contrÃ¡rio, se a chave for nova, adiciona ao dicionÃ¡rio e define a contagem como 1
            else:
                result[pair] = 1

    return result
````

## Treine seu Modelo usando Naive Bayes

Naive Bayes Ã© um algoritmo que pode ser usado para anÃ¡lise de sentimento. Ele leva pouco tempo para treinar e tambÃ©m tem um tempo de previsÃ£o curto.

#### EntÃ£o, como vocÃª treina um classificador de Naive Bayes?
- A primeira parte do treinamento de um classificador de Naive Bayes Ã© identificar o nÃºmero de classes que vocÃª tem.
- VocÃª criarÃ¡ uma probabilidade para cada classe.
$P(D_{pos})$ Ã© a probabilidade de que o documento seja positivo.
$P(D_{neg})$ Ã© a probabilidade de que o documento seja negativo.
Use as fÃ³rmulas a seguir e armazene os valores em um dicionÃ¡rio:

$$P(D_{pos}) = \frac{D_{pos}}{D}\tag{1}$$

$$P(D_{neg}) = \frac{D_{neg}}{D}\tag{2}$$

Onde $D$ Ã© o nÃºmero total de documentos, ou tweets neste caso, $D_{pos}$ Ã© o nÃºmero total de tweets positivos e $D_{neg}$ Ã© o nÃºmero total de tweets negativos.

#### Prior e Logprior

A probabilidade a priori representa a probabilidade subjacente na populaÃ§Ã£o-alvo de que um tweet seja positivo versus negativo. Em outras palavras, se nÃ£o tivÃ©ssemos nenhuma informaÃ§Ã£o especÃ­fica e escolhÃªssemos cegamente um tweet do conjunto populacional, qual seria a probabilidade de ele ser positivo versus negativo? Isso Ã© o "prior".

O prior Ã© a razÃ£o das probabilidades $\frac{P(D_{pos})}{P(D_{neg})}$. Podemos pegar o log do prior para reescalÃ¡-lo, e chamaremos isso de logprior.

$\text{logprior} = \log \left( \frac{P(D_{pos})}{P(D_{neg})} \right) = \log \left( \frac{D_{pos}}{D_{neg}} \right)$.

Observe que $\log(\frac{A}{B})$ Ã© o mesmo que $\log(A) - \log(B)$. EntÃ£o, o logprior tambÃ©m pode ser calculado como a diferenÃ§a entre dois logs:

$$\text{logprior} = \log (P(D_{pos})) - \log (P(D_{neg})) = \log (D_{pos}) - \log (D_{neg})\tag{3}$$


#### Probabilidade Positiva e Negativa de uma Palavra
Para calcular a probabilidade positiva e a probabilidade negativa para uma palavra especÃ­fica no vocabulÃ¡rio, usaremos as seguintes entradas:

- $freq_{pos}$ e $freq_{neg}$ sÃ£o as frequÃªncias dessa palavra especÃ­fica na classe positiva ou negativa. Em outras palavras, a frequÃªncia positiva de uma palavra Ã© o nÃºmero de vezes que a palavra Ã© contada com o rÃ³tulo 1.
- $N_{pos}$ e $N_{neg}$ sÃ£o o nÃºmero total de palavras positivas e negativas para todos os documentos (para todos os tweets), respectivamente.
- $V$ Ã© o nÃºmero de palavras Ãºnicas em todo o conjunto de documentos, para todas as classes, sejam elas positivas ou negativas.

Usaremos esses valores para calcular a probabilidade positiva e a probabilidade negativa para uma palavra especÃ­fica usando esta fÃ³rmula:

$$ P(W_{pos}) = \frac{freq_{pos} + 1}{N_{pos} + V}\tag{4} $$
$$ P(W_{neg}) = \frac{freq_{neg} + 1}{N_{neg} + V}\tag{5} $$

Observe que adicionamos "+1" no numerador para suavizaÃ§Ã£o aditiva. Este [artigo da wiki](https://en.wikipedia.org/wiki/Additive_smoothing) explica mais sobre suavizaÃ§Ã£o aditiva.

#### Log likelihood
Para calcular a logverossimilhanÃ§a dessa mesma palavra, podemos implementar as seguintes equaÃ§Ãµes:

$$\text{loglikelihood} = \log \left(\frac{P(W_{pos})}{P(W_{neg})} \right)\tag{6}$$

````python
# DicionÃ¡rio para armazenar a contagem de palavras
freqs = count_tweets({}, train_x, train_y)
````

````python
def train_naive_bayes(freqs, train_x, train_y):
    '''
    Input:
        freqs: dicionÃ¡rio de (palavra, rÃ³tulo) para a frequÃªncia da palavra
        train_x: uma lista de tweets
        train_y: uma lista de rÃ³tulos correspondentes aos tweets (0,1)
    Output:
        logprior: o logaritmo da probabilidade a priori
        loglikelihood: o logaritmo da verossimilhanÃ§a da equaÃ§Ã£o de Naive Bayes
    '''
    loglikelihood = {}
    logprior = 0

    # calcular V, o nÃºmero de palavras Ãºnicas no vocabulÃ¡rio
    vocab = set([pair[0] for pair in freqs.keys()])
    V = len(vocab)

    # calcular N_pos, N_neg
    N_pos = sum([freqs.get((word, 1), 0) for word in vocab])
    N_neg = sum([freqs.get((word, 0), 0) for word in vocab])

    # Calcular D, o nÃºmero de documentos
    D = len(train_x)

    # Calcular D_pos, o nÃºmero de documentos positivos
    D_pos = sum(train_y)

    # Calcular D_neg, o nÃºmero de documentos negativos
    D_neg = D - D_pos

    # Calcular logprior
    logprior = np.log(D_pos) - np.log(D_neg)

    # Para cada palavra no vocabulÃ¡rio...
    for word in vocab:
        # obter a frequÃªncia positiva e negativa da palavra
        freq_pos = freqs.get((word, 1), 0)
        freq_neg = freqs.get((word, 0), 0)

        # calcular a probabilidade de que cada palavra seja positiva e negativa
        p_w_pos = (freq_pos + 1) / (N_pos + V)
        p_w_neg = (freq_neg + 1) / (N_neg + V)

        # calcular o logaritmo da verossimilhanÃ§a da palavra
        loglikelihood[word] = np.log(p_w_pos / p_w_neg)

    return logprior, loglikelihood
````

## Teste seu Naive Bayes

Agora que temos o logprior e a loglikelihood, podemos testar a funÃ§Ã£o Naive Bayes fazendo previsÃµes em alguns tweets!

### naive_bayes_predict

**InstruÃ§Ãµes**:

* A funÃ§Ã£o recebe um tweet, logprior, e loglikelihood.
* Ela retorna a probabilidade de que o tweet pertenÃ§a Ã  classe positiva ou negativa.
* Para cada tweet, some as log-verossimilhanÃ§as de cada palavra no tweet.
* Adicione tambÃ©m o logprior a essa soma para obter o sentimento previsto desse tweet.

$$ p = \text{logprior} + \sum_i^N (\text{loglikelihood}_i)$$

#### Nota
Observe que calculamos o prior a partir dos dados de treinamento, e que os dados de treinamento estÃ£o igualmente divididos entre rÃ³tulos positivos e negativos (4000 tweets positivos e 4000 tweets negativos). Isso significa que a razÃ£o de positivo para negativo Ã© 1, e o logprior Ã© 0.

O valor de 0.0 significa que, quando adicionamos o logprior Ã  loglikelihood, estamos apenas adicionando zero Ã  loglikelihood. No entanto, lembre-se de incluir o logprior, porque sempre que os dados nÃ£o estiverem perfeitamente balanceados, o logprior serÃ¡ um valor diferente de zero.

````python
def naive_bayes_predict(tweet, logprior, loglikelihood):
    '''
    Input:
        tweet: uma string
        logprior: um nÃºmero
        loglikelihood: um dicionÃ¡rio de palavras mapeando para nÃºmeros
    Output:
        p: a soma de todas as logverossimilhanÃ§as de cada palavra no tweet (se encontradas no dicionÃ¡rio) + logprior (um nÃºmero)

    '''
  
    # processa o tweet para obter uma lista de palavras
    word_l = process_tweet(tweet)

    # inicializa a probabilidade para zero
    p = 0

    # adiciona o logprior
    p += logprior

    for word in word_l:

        # verifica se a palavra existe no dicionÃ¡rio de logverossimilhanÃ§as
        if word in loglikelihood:
            # adiciona a logverossimilhanÃ§a dessa palavra Ã  probabilidade
            p += loglikelihood[word]


    return p
````

### test_naive_bayes
Implemente test_naive_bayes.

**InstruÃ§Ãµes**:
* Implemente `test_naive_bayes` para verificar a precisÃ£o de suas previsÃµes.
* A funÃ§Ã£o recebe `test_x`, `test_y`, log_prior, e loglikelihood.
* Ela retorna a precisÃ£o do seu modelo.
* Primeiro, use a funÃ§Ã£o `naive_bayes_predict` para fazer previsÃµes para cada tweet em test_x.

````python
def test_naive_bayes(test_x, test_y, logprior, loglikelihood, naive_bayes_predict=naive_bayes_predict):
    """
    Input:
        test_x: Uma lista de tweets
        test_y: os rÃ³tulos correspondentes para a lista de tweets
        logprior: o logprior
        loglikelihood: um dicionÃ¡rio com as logverossimilhanÃ§as para cada palavra
    Output:
        accuracy: (# de tweets classificados corretamente)/(total # de tweets)
    """
    accuracy = 0  # retorne isso corretamente

    # lista para armazenar os valores previstos
    y_hats = [naive_bayes_predict(tweet, logprior, loglikelihood) for tweet in test_x]
    
    # 0 se o valor previsto for negativo e 1 se o valor previsto for positivo
    y_hats_binary = [1 if y_hat > 0 else 0 for y_hat in y_hats]

    # calcular a quantidade de valores previstos corretamente
    correct_predictions = np.sum(np.array(y_hats_binary) == np.array(test_y))

    # calcular a precisÃ£o
    accuracy = correct_predictions / len(test_y)

    return accuracy
````

## Filtrar palavras por RazÃ£o de Contagens Positivas para Negativas

- Algumas palavras tÃªm mais contagens positivas do que outras e podem ser consideradas "mais positivas". Da mesma forma, algumas palavras podem ser consideradas mais negativas do que outras.
- Uma maneira de definir o nÃ­vel de positividade ou negatividade, sem calcular a verossimilhanÃ§a, Ã© comparar a frequÃªncia positiva com a frequÃªncia negativa da palavra.
    - Observe que tambÃ©m podemos usar os cÃ¡lculos de verossimilhanÃ§a para comparar a positividade ou negatividade relativa das palavras.
- Podemos calcular a razÃ£o de frequÃªncias positivas para negativas de uma palavra.
- Assim que conseguirmos calcular essas razÃµes, tambÃ©m podemos filtrar um subconjunto de palavras que tenham uma razÃ£o mÃ­nima de positividade/negatividade ou maior.
- Da mesma forma, tambÃ©m podemos filtrar um subconjunto de palavras que tenham uma razÃ£o mÃ¡xima de positividade/negatividade ou menor (palavras que sÃ£o pelo menos tÃ£o negativas ou atÃ© mais negativas do que um determinado limite).


### get_ratio

- Dado o dicionÃ¡rio freqs de palavras e uma palavra em particular, use `lookup(freqs,word,1)` para obter a contagem positiva da palavra.
- Da mesma forma, use a funÃ§Ã£o `lookup` para obter a contagem negativa dessa palavra.
- Calcule a razÃ£o das contagens positivas divididas pelas contagens negativas

 $$ratio = \frac{\text{pos\_words} + 1}{\text{neg\_words} + 1}$$

Onde pos_words e neg_words correspondem Ã  frequÃªncia das palavras em suas respectivas classes.

````python
def lookup(freqs, word, label):
    '''
    Entrada:
        freqs: um dicionÃ¡rio com a frequÃªncia de cada par (ou tupla)
        word: a palavra a ser procurada
        label: o rÃ³tulo correspondente Ã  palavra
    SaÃ­da:
        n: o nÃºmero de vezes que a palavra com seu rÃ³tulo correspondente aparece.
    '''
    n = 0  # freqs.get((word, label), 0)

    pair = (word, label)
    if pair in freqs:
        n = freqs[pair]

    return n
````

````python
def get_ratio(freqs, word):
    '''
    Entrada:
        freqs: dicionÃ¡rio contendo as palavras

    SaÃ­da: um dicionÃ¡rio com as chaves 'positive', 'negative' e 'ratio'.
        Exemplo: {'positive': 10, 'negative': 20, 'ratio': 0.5}
    '''
    pos_neg_ratio = {'positive': 0, 'negative': 0, 'ratio': 0.0}
    
    # use lookup() para encontrar as contagens positivas para a palavra (denotada pelo inteiro 1)
    pos_neg_ratio['positive'] = lookup(freqs, word, 1)
    
    # use lookup() para encontrar as contagens negativas para a palavra (denotada pelo inteiro 0)
    pos_neg_ratio['negative'] = lookup(freqs, word, 0)
    
    # calcular a razÃ£o de contagens positivas para negativas para a palavra
    pos_neg_ratio['ratio'] = (pos_neg_ratio['positive'] + 1) / (pos_neg_ratio['negative'] + 1)
  
    return pos_neg_ratio
````

### get_words_by_threshold

* Se definirmos o rÃ³tulo como 1, procuraremos todas as palavras cujo limiar de positivo/negativo seja pelo menos tÃ£o alto quanto esse limiar ou maior.
* Se definirmos o rÃ³tulo como 0, procuraremos todas as palavras cujo limiar de positivo/negativo seja no mÃ¡ximo tÃ£o baixo quanto o limite dado ou menor.
* Use a funÃ§Ã£o `get_ratio` para obter um dicionÃ¡rio contendo a contagem positiva, contagem negativa e a razÃ£o das contagens positivas para negativas.
* Anexe o dicionÃ¡rio `get_ratio` dentro de outro dicionÃ¡rio, onde a chave Ã© a palavra e o valor Ã© o dicionÃ¡rio `pos_neg_ratio` que Ã© retornado pela funÃ§Ã£o `get_ratio`.
Um par chave-valor de exemplo teria esta estrutura:
```
{'happi':
    {'positive': 10, 'negative': 20, 'ratio': 0.524}
}
```
```

````python
def get_words_by_threshold(freqs, label, threshold, get_ratio=get_ratio):
    '''
    Entrada:
        freqs: dicionÃ¡rio de palavras
        label: 1 para positivo, 0 para negativo
        threshold: razÃ£o que serÃ¡ usada como limite para incluir uma palavra no dicionÃ¡rio retornado
    SaÃ­da:
        word_list: dicionÃ¡rio contendo a palavra e informaÃ§Ãµes sobre sua contagem positiva, contagem negativa e razÃ£o de contagens positivas para negativas.
        exemplo de um par chave-valor:
        {'happi':
            {'positive': 10, 'negative': 20, 'ratio': 0.5}
        }
    '''
    word_list = {}
    for key in freqs.keys():
        word, _ = key

        # obter a razÃ£o positiva/negativa para uma palavra
        pos_neg_ratio = get_ratio(freqs, word)

        # se o rÃ³tulo for 1 e a razÃ£o for maior ou igual ao limite...
        if label == 1 and pos_neg_ratio['ratio'] >= threshold:
        
            # Adicionar a razÃ£o positiva/negativa ao dicionÃ¡rio
            word_list[word] = pos_neg_ratio

        # Se o rÃ³tulo for 0 e a razÃ£o positiva/negativa for menor ou igual ao limite...
        elif label == 0 and pos_neg_ratio['ratio'] <= threshold:
        
            # Adicionar a razÃ£o positiva/negativa ao dicionÃ¡rio
            word_list[word] = pos_neg_ratio

        # caso contrÃ¡rio, nÃ£o incluir essa palavra na lista (nÃ£o fazer nada)

    return word_list
````
